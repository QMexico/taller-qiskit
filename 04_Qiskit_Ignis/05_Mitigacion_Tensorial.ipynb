# Introducción

Los errores de medición que observamos en el notebook anterior ([04_Errores_de_Medición](04_Errores_de_Medicion.ipynb)) son específicos del caso cuando el mismo error afecta a todos los qubits de un dispositivo. Sin embargo, puede suceder que sólamente algunos de los qubits estén sujetos a errores de medición. En este caso se dice que el error puede ser localizado en un subconjunto de qubits. Así, aún cuando nuestro dispositivo tenga $n$ qubits, serán necesarios menos de $2^n$ circuitos de calibración para corregir dichos errores. Al procedimiento que corrige los errores de medición cuando son locales se le conoce como __Mitigación Tensorial (MT)__.

```python
# Herramientas de Python auxiliares a Qiskit
import numpy as np
import matplotlib.pyplot as plt

# Herramientas de Qiskit necesarias para crear y ejecutar circuitos cuánticos en simulador y dispositivos
import qiskit
from qiskit import QuantumRegister, QuantumCircuit, ClassicalRegister, Aer, execute
from qiskit.providers.aer import noise
from qiskit.providers.ibmq import least_busy
from qiskit.tools.monitor import job_monitor
from qiskit.tools.visualization import plot_histogram

# Herramientas de Qiskit necesarias para crear los modelos de ruido locales y globales
from qiskit.providers.aer.noise import NoiseModel
from qiskit.providers.aer.noise.errors import readout_error, pauli_error

# Herramientas de Qiskit Ignis que utilizaremos
from qiskit.ignis.mitigation.measurement import (tensored_meas_cal,
                                                 TensoredMeasFitter)

```

<a id="intro1"></a> 
## Modelos de ruido locales

Para comenzar con la mitigación tensorial debemos definir un modelo de ruido que afecte a algunos subconjuntos de qubits de manera distinta. Por ejemplo, consideremos un dispositivo de tres qubits donde cada uno tiene un modelo de error distinto. En particular, consideremos que solo el segudo de nuestros qubits está afectado por ruido:


```python
# Modelo de ruido local
##########################
local_noise_model = NoiseModel()

# Error para un solo qubit
# Los valores representan las entradas de la matriz de calibración! Trataremos de verificarlo
# [P(0|0), P(0|1)]
# [P(1|0), P(1|1)]

read_err = noise.errors.readout_error.ReadoutError([[1,0],[0,1]]) # Prueba modificando los valores
read_err_2 = noise.errors.readout_error.ReadoutError([[0.6, 0.4],[0.4,0.6]]) # Prueba modificando los valores
read_err_3 = noise.errors.readout_error.ReadoutError([[1,0],[0,1]]) # Prueba modificando los valores

# Agregamos el modelo de error de medición los tres
local_noise_model.add_readout_error(read_err, [0])
local_noise_model.add_readout_error(read_err_2, [1])
local_noise_model.add_readout_error(read_err_3, [2])
##########################
```

Para observar nuestro modelo de ruido local ejecutaremos un circuito de prueba, lo ejecutaremos en un simulador sin ruido. Nuevamente veremos errores en las mediciones. La naturaleza de los errores variará mucho dependiendo de qué qubit(s) esté(n) siendo afectado(s) y en qué medida. 

```python
#Creamos un circuito de prueba que ejecutaremos en un simulador y en un dispositivo real
qreg = QuantumRegister(3)
creg = ClassicalRegister(3)
qc = QuantumCircuit(qreg,creg)
qc.h(0)
qc.cx(0,1)
qc.cx(1,2)
qc.barrier()
qc.measure(range(0,3),range(0,3))
 
simulator = Aer.get_backend("qasm_simulator")

# Ejecución en simulador sin ruido
sim_counts = execute(qc, backend = simulator, shots = 1024, optimization_level = 0).result().get_counts()

# Ejecución en simulador con ruido local
local_noise_counts = execute(qc, backend = simulator, shots = 1024, optimization_level = 0, 
                           noise_model = local_noise_model).result().get_counts()
```

```python
# Comparación de conteos en los dispositivos con ruido general, local, y sin ruido
plot_histogram([sim_counts,local_noise_counts], legend = ["Sin ruido","Ruido local"])
```

Podemos observar en nuestra gráfica de conteos que la aparición de errores sólo se debe a las variaciones de medición en el segundo qubit, no hay conteos erróneos debido a los primer y tercer qubits. __Puedes modificar el modelo de error para observar las variaciones en los conteos.__

# Circuitos de calibración para Mitigación Tensorial

Al considerar que los errores suceden de manera local, serán necesarios menos circuitos de calibración. Si se tienen $m$ qubits decritos por un mismo modelo de ruido, es decir que el error es local a $m$ qubits, entonces serán necesarios $2^m$ circuitos de calibración. 

Es importante recalcar que un modelo de ruido puede ser local en más de un qubit, por ejemplo, cuando dos de tres qubits tienen el mismo modelo de ruido, y otro de ellos tiene un modelo de ruido invididual. En este caso, el número de circuitos necesarios para la calibración será $2^2 = 4$. Esto se debe a que en la mitigación tensorial solo es necesario que aparezcan al menos una vez los estados base de $m$ qubits. Esto se satisface si se generan cuatro circuitos que generen 

$$\ket{000},\ket{010},\ket{110},\ket{111}$$


En nuestro caso $m = 1$ y solo serán necesarios dos circuitos de calibración que generen

$$ \ket{000}\  \text{y}\  \ket{111}$$


<a id="secc1.2"></a>
## Cómo construir circuitos de calibración para MT en Qiskit


### Función:  `tensored_meas_cal(mit_pattern, qr, cr, circlabel)`:
> __¿Qué hace?__: Sea $m$ el mayor número de qubits descrito por un modelo de ruido local, entonces se regresan $2^m$ circuitos de calibración que forman la mínima cantidad de circuitos necesarios para llevar a cabo la caracterización.
    
> __¿Qué argumentos toma?__ 
- ``mit_pattern (Optional[List[List[int]]])``: Qubits a los que se les aplicará la corrección, separados en los grupos que indican cuales están descritos por un mismo modelo de ruido local.
- ``qr (Union[int, List[QuantumRegister], None])``: Nombre o tamaño del registro cuántico a utilizar.
- ``cr (Union[int, List[ClassicalRegister], None])``: Nombre o tamaño del registro clásico a utilizar.
- ``circlabel (str)``: Una cadena de caracteres que se agregará al nombre de cada circuito para identificarlos.
    
> __¿Qué devuelve?__: Devuelve una tupla con dos listas. La primera de las listas está compuesta de $2^m$ objetos `QuantumCircuit` que generan cada uno de los estados de la base, mientras que la segunda lista está compuesta de `str`, cadenas que identifican a cada circuito de calibración.

```python
# Mitigación Tensorial

#####################
# tensored_meas_cal #
#####################
# Indicamos que cada qubit tiene su modelo de ruido independiente con mit_pattern = [[0],[1],[2]], 
# si en alguna sub-lista hubiera más de un número indicaría que esos dos qubits comparten modelo de ruido

tens_calcircs, tens_state_labels = tensored_meas_cal( mit_pattern = [[0],[1],[2]], qr = qreg, circlabel = "t") 

# Para verificar cuántos circuitos de calibración tenemos
for circ in tens_calcircs:
    print(circ.name)
```

<a id="secc2"></a>
# Matrices de Calibración 

La calibración tensorial produce una matriz de calibración por cada conjunto de qubits que esté descrito por un mismo modelo de ruido.


### Clase:  `TensoredMeasFitter( results, mit_pattern, substate_labels_list, circlabel)`:
> __¿Qué hace?__: Una vez que se ejecutaron los circuitos de calibración en un dispositivo ruidoso, esta clase permite tomar los resultados y generar un objeto que contiene una matriz de calibración y un filtro de calibración. Este objeto es útil para conocer completamente y, posteriormente, corregir los errores.
    
> __¿Qué parámetros de construcción tiene?__ 
- ``results (Union[Result, List[Result]])``: Los resultados que se obtienen al correr los circuitos de calibración generados por `tensored_meas_cal()`.
- ``mit_pattern (List[List[int]])``: Qubits a los que se les aplicará la corrección, separados en los grupos que indican cuáles están descritos por un mismo modelo de ruido local.
- ``substate_labels_list (Optional[List[int]]) ``: Las etiquetas que aparecerán en las filas y columnas de cada matriz de calibración.
- ``circlabel (str)``: Una cadena de caracteres que se agregará al nombre de cada circuito para identificarlos.
    
> __¿Qué construye?__: Construye un objeto que contiene dos atributos principales. Primero inicializa una o más matrices de calibración a partir del resultado de ejecutar los circuitos de calibración. Se tendrán $n$ circuitos de calibración si se tienen $n$ modelos de ruido local. Se puede acceder a las matrices $n$ de calibración haciendo ``tens_matcal = tensfitter.cal_matrices(i)`` con el índice $i \in {0,1,...,n}$. A partir de estas matrices de calibración se generan filtros, `tens_filter = tensfitter.filter` que permiten corregir los errores de medición causados por el modelo de ruido. Ese filtro se aplica a los resultados de un circuito ejecutado con errores mediante, `result_tenscorrected = tens_filter.apply(locnoise_results)`


<a id="secc2.1"></a>
## Obteniendo las matrices de calibración en Qiskit

```python
# Comenzamos por ejecutar cada uno de los circuitos de calibración en un backend con nuestro modelo de
# tres ruidos locales, uno para cada qubit

locnoise_results = execute(tens_calcircs, backend = simulator, shots = 1024, optimization_level = 0, 
                           noise_model = local_noise_model).result()


######################
# TensoredMeasFitter #
######################


tens_fitter = TensoredMeasFitter(locnoise_results, mit_pattern=[[0],[1],[2]])

print('Matriz de calibración para Q0')
tens_fitter.plot_calibration(0)
print('Matriz de calibración para Q1')
tens_fitter.plot_calibration(1)
print('Matriz de calibración para Q2')
tens_fitter.plot_calibration(2)
```

Tal y como era de esperarse, según cómo se definió en el modelo de ruido de cada qubit, solo el segundo qubit se ve afectado por ruido (de forma muy severa). Notemos que los valores en cada cuadrante de las matrices de densidad corresponden a los valores indicados en nuestro modelo. __Puedes modificar el modelo de error para observar las variaciones en las entradas de cada matriz de densidad.__ 

Podemos obtener la [fidelidad de medición](https://www.fisicalab.com/apartado/medidas-precision#fidelidad) de cada uno de los qubits de manera sencilla.

```python

print("La fidelidad de medición del qubit Q0 es:", tens_fitter.readout_fidelity(0))

print("La fidelidad de medición del qubit Q1 es:", tens_fitter.readout_fidelity(1))

print("La fidelidad de medición del qubit Q2 es:", tens_fitter.readout_fidelity(2))
```

<a id="secc3"></a>
# Mitigando errores locales

Finalmente podemos arreglar los resultados de un experimento afectado por un modelo de ruido local aplicando la información almacenada en cada una de nuestras matrices de densidad

```python
# Get the filter object
tens_filter = tens_fitter.filter

# Nuevamente ejecutamos el circuito con el modelo de ruido local, 
# almacenamos su resultado y accedemos a sus conteos
locnoise_results = execute(qc, backend = simulator, shots = 1024, optimization_level = 0, 
                           noise_model = local_noise_model).result()

locnoise_counts = locnoise_results.get_counts()

# Corregimos los resultados con el filtro y accedemos a sus conteos
tenscorrected_results = tens_filter.apply(locnoise_results)
tenscorrected_counts = tenscorrected_results.get_counts()
```

```python
# Comparación de conteos: Ideal, con ruido local, con Mitigación Tensorial
plot_histogram([sim_counts, locnoise_counts, tenscorrected_counts], 
               legend = ["Sin ruido","Con ruido local", "Mitigado"])
```

Nuevamente observamos que la mitigación tensorial funciona bien para recuperar resultados similares a un caso ideal.
